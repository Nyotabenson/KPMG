```python
import pandas as pd
```

```python
# files
dataset = "KPMG_VI_New_raw_data_update_final.xlsx"
sheet1 = "Transactions"
sheet2 = "NewCustomerList"
sheet3 = "CustomerDemographic"
sheet4 = "CustomerAddress"
```

 <h1 style="font-size:8rem;color:red;"> Transactions</h1>

```python
Transactions = pd.read_excel(dataset, sheet_name=sheet1)
```

```python
Transactions.head()
```

```python
Transactions.shape
```

```python
Transactions.isna().sum()
```

```python
Transactions.online_order.value_counts() 
```

```python
# fill in the missing data
Transactions["online_order"].fillna(0, inplace=True)
missing_list = ["brand", "product_line", "product_class", "product_size"]
for i in missing_list:
    Transactions[i].fillna("missing", inplace=True)
    
Transactions["standard_cost"].fillna(Transactions["standard_cost"].mean(), inplace=True)
Transactions["product_first_sold_date"].fillna(Transactions["product_first_sold_date"].mean(), inplace=True)
```

```python
Transactions.describe()
```

```python
Transactions.sort_values("transaction_date")
```

```python
Transactions['transaction_date'].min()
```

```python
unique = ["product_line","product_class","product_size"]
for i in unique:
    print(Transactions[i].unique())
```

 <h1 style="font-size:5rem;color:red;"> NewCustomerList</h1>

```python
NewCustomerList = pd.read_excel(dataset, sheet_name=sheet2).drop(columns=
                            ["Unnamed: 16","Unnamed: 17","Unnamed: 18","Unnamed: 19","Unnamed: 20"], axis=1)
```

```python
pd.set_option('display.max_columns', None)
NewCustomerList.head()
```

```python
NewCustomerList.shape
```

```python
NewCustomerList.isna().sum()
```

```python
missing = ["last_name", "job_title", "job_industry_category"]
for i in missing:
    NewCustomerList[i].fillna("missing", inplace=True)
NewCustomerList.dropna(inplace=True)
```

```python
NewCustomerList.isna().sum()
```

```python
NewCustomerList.describe()
```

```python
NewCustomerList.sort_values("DOB").tail()
```

```python
unique=["gender", "wealth_segment", "deceased_indicator", "owns_car"]
for i in unique:
    print(NewCustomerList[i].unique())
```

 <h1 style="font-size:5rem;color:red;"> CustomerDemographic</h1>

```python
CustomerDemographic = pd.read_excel(dataset, sheet_name=sheet3)
```

```python
CustomerDemographic.head()
```

```python
CustomerDemographic.shape
```

```python
CustomerDemographic.isna().sum()
```

```python
CustomerDemographic=CustomerDemographic.drop(columns=["default"], axis=1)

missing = ["last_name", "job_title", "job_industry_category"]
for i in missing:
    CustomerDemographic[i].fillna("missing", inplace=True)
    
CustomerDemographic['tenure'].fillna(CustomerDemographic['tenure'].mean(), inplace=True)

CustomerDemographic.dropna(inplace=True)
```

```python
CustomerDemographic.isna().sum()
```

```python
CustomerDemographic.describe()
```

```python
CustomerDemographic = CustomerDemographic.drop(CustomerDemographic.index[33])
```
