```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
custdem = pd.read_csv("CustomerDemographic_eng.csv")
custdem.head()
```

```python
custdem.shape
```

```python
custdem.columns
```

```python
custdemmodel1 = custdem.drop(['customer_id', 'first_name', 'last_name', 'DOB','deceased_indicator'], axis=1)
custdemmodel1.head()
```

```python
custdemmodel1.describe()
```

```python
# Binning the continuous data
custdemmodel1['bike_purchases'] = pd.cut(x=custdemmodel1['past_3_years_bike_related_purchases'],
                                  bins=[0,48,99],
                                  labels=['High', 'Low'])
custdemmodel1['tenure_c'] = pd.cut(x=custdemmodel1['tenure'],
                          bins=[0,11,22],
                          labels=['Long', 'Short'])
custdemmodel1.head()
```

```python
custdemmodel2 = custdemmodel1.drop(['past_3_years_bike_related_purchases', 'tenure', 'age'], axis=1)
custdemmodel2
```

```python
custdemmodel2.dtypes
```

```python
for column in custdemmodel2.columns:
    if custdemmodel2[column].dtype=='object':
        custdemmodel2[column] = custdemmodel2[column].astype('category')   
```

```python
custdemmodel2.dtypes
```

```python
#train and test to avoid data leakage
X = custdemmodel2.drop('Age_group', axis=1)
y = custdemmodel2['Age_group']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```
